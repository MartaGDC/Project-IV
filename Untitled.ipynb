```python
import requests
from bs4 import BeautifulSoup
import datetime
```

STEAMCHARTS is one of the available websites to monitor the evolution of users and consumption for games, and the only one free and available for webscrapping.

Other webs considered were VG Insights and Steam DB.

```python
# Webscrapping of the main page of steamcharts.com/top -> Maybe add more later
url = "https://steamcharts.com/top" # https://steamcharts.com/top/p.2 and https://steamcharts.com/top/p.3...
res = requests.get(url)
html = res.content
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
name_game_unpross = soup.find_all("a")
name_game = [name_game_unpross[i].getText().strip() for i in range(len(name_game_unpross))]
name_game
```

```python
name_game = name_game[3 :]
name_game = name_game[: -6]
```

```python
len(name_game)
```

```python
player_hours_unpross = soup.find_all("td", {"class": "num period-col player-hours"})
player_hours = [i.getText() for i in player_hours_unpross]
player_hours
```

```python
len(player_hours)
```

```python
peak_players_unpross = soup.find_all("td", {"class": "num period-col peak-concurrent"})
peak_players = [i.getText() for i in peak_players_unpross]
peak_players
```

```python
len(peak_players)
```

```python
numbers_unpross = soup.find_all("td", {"class": "num"})
numbers = [i.getText() for i in numbers_unpross]
numbers
```

```python
current_players = [numbers[i] for i in range(len(numbers)) if (i+1)%3 == 1]
current_players
```
